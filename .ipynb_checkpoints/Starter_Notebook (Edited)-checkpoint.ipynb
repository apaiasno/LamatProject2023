This is a starter notebook for learning how to use the routines in SPEX-PRISM_lib.py to read in spectra from the SpeX Prism library.

# Prerequisites
To use this notebook, you must download the SpeX Prism library to a folder called SPEX-PRISM that lives in the same directory as this notebook.

1. Go to: https://download-directory.github.io/  

This a resource for downloading a specific folder in a Github repository. 

2. Enter this link in the box: https://github.com/aburgasser/splat/tree/master/resources/Spectra/SPEX-PRISM

This is a specific folder in a Github repository for a tool called "splat" that contains many FITS files of spectra that make up the SPEX-PRISM library. FITS is a file type commonly used in astronomy used to save data; you can read more about it here if you are interested: https://en.wikipedia.org/wiki/FITS/  

3. After the zipped folder has downloaded, unzip it in your project home directory. Name the folder of FITS files "SPEX-PRISM". 

If you are on a Linux/UNIX-based system, you can do this in the terminal by moving the zipped folder (suppose the name of it is "spex-prism.zip") to your project home directory and running the command: 
unzip spex-prism.zip -d SPEX-PRISM 

# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import SPEX_PRISM_lib as SPL
```

# List all available spectral types in SpeX Prism library

```python
SpT_all = SPL.list_all_available_SpT()
[print(SpT) for SpT in SpT_all];
```

# Plot all spectra of a given spectral type in the SpeX Prism library

```python
SpT = 'T6'
wavs, fluxes, errs = SPL.get_all_SpT(SpT)

plt.figure(figsize=(10,6))
for i in range(len(wavs)):
    plt.errorbar(wavs[i], fluxes[i], abs(errs[i]), fmt='.')

plt.xlabel('Wavelength (microns)')
plt.ylabel('Flux')
plt.show()
```

## For M8:

```python
filename = '10014_10817.fits' # first spectrum listed in SpeX Prism library

wav, flux, err, SpT = SPL.read_spectrum(filename)
plt.figure(figsize=(10,6))
plt.errorbar(wav, flux, abs(err), fmt='-')
plt.xlabel('Wavelength (microns)')
plt.ylabel('Flux')
Carbon_monoxide = ("CO")
Iron_hydride = ("FeH")
Methane = ("CH4")
Water = ("H20")

plt.text(1.15, 0.8, Water, fontsize = 8,)
plt.text(1.48, 0.55, Water, fontsize = 8)
plt.text (2.3, 0.38, Carbon_monoxide, fontsize = 8)
plt.text (1, 0.7, Iron_hydride, fontsize = 8)
plt.text (1.17, 0.9, Iron_hydride, fontsize = 8)
plt.text (2.2275, 0.275, Methane, fontsize = 8)

#plt.legend(loc="upper left")
```

```python
T6 = '10004_10916.fits'
M8 = '10014_10817.fits'

wavT6, fluxT6, errT6, SpTT6 = SPL.read_spectrum(T6)
wavM8, fluxM8, errM8, SpTM8 = SPL.read_spectrum(M8)
plt.figure(figsize=(10,6))
plt.errorbar(wavT6, fluxT6, abs(errT6), fmt='-', label = "T6")
plt.errorbar(wavM8, fluxM8, abs(errM8), fmt='-', label = "M8")
plt.axvspan(.98, 1.03, color = "lightgrey", label = "FeH")
plt.xlabel('Wavelength (microns)')
plt.ylabel('Flux')
plt.title("T6 vs. M8")
plt.grid()
plt.xlim(0.8,1.5)
plt.legend(loc="upper left")
plt.show()
```
