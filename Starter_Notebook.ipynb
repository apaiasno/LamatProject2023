FITS is a file type commonly used in astronomy used to save data; you can read more about it here if you are interested: https://en.wikipedia.org/wiki/FITS/  

3. After the zipped folder has downloaded, unzip it in your project home directory. Name the folder of FITS files "SPEX-PRISM". 

If you are on a Linux/UNIX-based system, you can do this in the terminal by moving the zipped folder (suppose the name of it is "spex-prism.zip") to your project home directory and running the command:

# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import SPEX_PRISM_lib as SPL
```

# List all available spectral types in SpeX Prism library

```python
SpT_all = SPL.list_all_available_SpT()
[print(SpT) for SpT in SpT_all];
```

# Plot all spectra of a given spectral type in the SpeX Prism library

```python
SpT = 'L0'
wavs, fluxes, errs = SPL.get_all_SpT(SpT)

plt.figure(figsize=(10,6))
for i in range(len(wavs)):
    plt.errorbar(wavs[i], fluxes[i], errs[i], fmt='.')

plt.xlabel('Wavelength (microns)')
plt.ylabel('Flux')
plt.show()
```
